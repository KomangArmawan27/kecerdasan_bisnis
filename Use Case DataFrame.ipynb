```python
import pandas as pd
from functools import reduce
```

```python
df_pupuk = pd.read_excel('Data Impor\Impor Pupuk.xls')
df_minyak = pd.read_excel('Data Impor\Import Minyak Bumi.xls')
df_sayur = pd.read_excel('Data Impor\Import Sayuran.xls')
```

```python
# menghilangkan nilai missing value
# df_sayur.dropna(axis=0)
```

# mencari tahu impor negara mesir dibidang pupuk, minyak, dan sayur

```python
# merging multiple tables
df_sayuran_minyak = pd.merge(df_sayur, df_minyak, 
                             left_on='Negara Asal', 
                             right_on='Negara Asal', 
                             how='left')
```

```python
# handel missing values
df_sayuran_minyak.fillna(0, inplace=True) # inplace = override
```

```python
# membersihkan kolom atau baris yang tidak dibutuhkan
df_sayuran_minyak_clean = df_sayuran_minyak[df_sayuran_minyak['Negara Asal']!=0]
```

```python
df_sayuran_minyak_clean[df_sayuran_minyak_clean['Negara Asal']=='Kanada']
```

```python
tables = df_sayur, df_minyak, df_pupuk
df_sayuran_minyak_pupuk = reduce(lambda  left,right: pd.merge(left,right,on=['Negara Asal'],
                                            how='outer'), tables)
```

```python
df_sayuran_minyak_pupuk.fillna(0, inplace = True)
df_sayuran_minyak_pupuk
```

```python
# making new variable for showing mesir only
df_sayuran_minyak_pupuk_mesir = df_sayuran_minyak_pupuk[df_sayuran_minyak_pupuk['Negara Asal']=='Mesir']
```
