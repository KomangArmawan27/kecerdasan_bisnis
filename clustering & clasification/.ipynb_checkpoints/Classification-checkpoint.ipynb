```python
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler # data preprocessing -> normalisasi data (pakai transformasi Z-Score)
from sklearn.model_selection import train_test_split # memecah data training dan data testing
from sklearn.metrics import mean_squared_error # ukuran akurasi
from sklearn.metrics import accuracy_score # ukuran akurasi
from sklearn.metrics import confusion_matrix # ukuran akurasi
from sklearn.metrics import classification_report # ukuran akurasi
from sklearn.model_selection import StratifiedKFold, cross_val_score,GridSearchCV # ukuran akurasi
from sklearn.preprocessing import LabelEncoder # data preprocessing -> discreatisasi
from datetime import datetime, timedelta, date # jika butuh data tanggal/waktu
import numpy as np # 
import matplotlib.pyplot as plt # visualisasi data
from sklearn.tree import DecisionTreeClassifier # import algoritma
from sklearn import tree
from sklearn.metrics import plot_confusion_matrix
```

```python
df = pd.read_csv('D:\kuliah\smt 7\kecerdasan bisnis\praktisi\Data Impor\diabetes.csv')
```

```python
# memisahkan variabel X dan y
# X = gejala/tanda - tanda diabetes
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[feature_cols]
y = df.Outcome
```

```python
X
```

```python
# Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# test_size = 0.30 artinya proporsi data testing adalah 30% dari total data
```

```python
X_train
```

```python
# modeling
# modeling menggunakan DEcisionTree
fit = DecisionTreeClassifier().fit(X_train, y_train.values.ravel())
y_pred = fit.predict(X_test) # prediction
```

```python
X_test['y_pred'] = y_pred
X_test['y_asli'] = y_test
X_test
```

```python
confusion_matriks = confusion_matrix(y_test, y_pred)
classification_reports = classification_report(y_test, y_pred)
```

```python
print(confusion_matriks)
```

```python
print(classification_reports)
```

```python
# variable importance
"""
Variabel mana yang paling penting dapat dilihat di bagian ini
"""
df_importance = pd.DataFrame(fit.feature_importances_).T
df_importance.columns = feature_cols
df_importance2 = df_importance.T.reset_index()
df_importance2.columns = ['variabel','relatif_importance']
df_importance2 = df_importance2.sort_values('relatif_importance',ascending=False)
df_importance2
```
